```python
#requirement 1.1
import time
def Task():
    print("Running Task")
    time.sleep(1)
t_start = time.perf_counter()
Task()
t_finish = time.perf_counter()
elapsed_seconds = t_finish - t_start
print("Elapsed time =",elapsed_seconds)
```

```python
#requiremenet 1.3
def Task1():
    print("Running Task 1")
    time.sleep(0.49)
def Task2():
    print("Running Task 2")
    time.sleep(0.49)
def Task3():
    print("Running Task 3")
    time.sleep(0.99)    

def run_frame(frame_index):
  if frame_index == 1: 
    print("starting frame 1")
    Task1()
    Task2()
    Task3()
  elif frame_index ==2: 
    print("starting frame 2")
    Task1()  
  elif frame_index ==3: 
    print("starting frame 3")
    Task1()
    Task2()
    Task3()    
  elif frame_index ==4: 
    print("starting frame 4")
    Task1()
  elif frame_index ==5: 
    print("starting frame 5")
    Task1()
    Task2() 
    Task3()
  elif frame_index ==6: 
    print("starting frame 6")
    Task1()
    Task2() 
  else:
    print("wrong frame number")
next = time.perf_counter()
frame_length = 2
minor_cycle = 1
first_minor_cycle = 1
last_minor_cycle = 6 #equals number of frames
while 1 : 
  while time.perf_counter() < next: 
    pass #wait
  print("current time:",time.perf_counter())
  run_frame(minor_cycle)
  next=next+frame_length
  if minor_cycle < last_minor_cycle : 
    minor_cycle = minor_cycle + 1
  else : 
    minor_cycle = first_minor_cycle #start major cycle again
```
